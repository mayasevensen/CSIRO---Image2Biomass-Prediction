The same image appears 4–5 times
But with different targets

Convert into one image with multiple targets

```python
import pandas as pd

df = pd.read_csv("../data/tabular/train/train.csv")

# Extract image_id from sample_id
df["image_id"] = df["sample_id"].str.split("__").str[0]

# Pivot targets to columns
df_wide = df.pivot_table(
    index=[
        "image_id",
        "image_path",
        "Sampling_Date",
        "State",
        "Species",
        "Pre_GSHH_NDVI",
        "Height_Ave_cm"
    ],
    columns="target_name",
    values="target"
).reset_index()

df_wide.head()
```

```python
df_wide.to_parquet("../data/processed/train_tabular.parquet")
```
